```python

def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

```

```python
customer_wrangled = customer_df.copy()
```

```python
for column in customer_wrangled.columns:
    if customer_wrangled[column].dtype in ['int64', 'float64']:  # Check if the column is numeric
        customer_wrangled = remove_outliers(customer_wrangled, column)
```

```python
customer_wrangled
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
threshold_unique_values = 20  # You can adjust this threshold based on your specific dataset
continuous_columns = [col for col in customer_wrangled.columns if customer_wrangled[col].nunique() > threshold_unique_values and customer_wrangled[col].dtype in ['int64', 'float64']]
```

```python
# One-hot encoding
one_hot_columns = ['state', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class']
customer_wrangled = pd.get_dummies(customer_wrangled, columns=one_hot_columns)

# Ordinal encoding
ordinal_mappings = {
    'coverage': {"Basic": 0, "Extended": 1, "Premium": 2},
    'employmentstatus': {"Employed": 2, "Unemployed": 0, "Medical Leave": 1, "Disabled": 1, "Retired": 1},
    'location_code': {"Rural": 0, "Suburban": 1, "Urban": 2},
    'vehicle_size': {"Small": 0, "Medsize": 1, "Large": 2}
}

for column, mapping in ordinal_mappings.items():
    customer_wrangled[column] = customer_wrangled[column].map(mapping)
```

```python
customer_wrangled['effective_to_date'] = pd.to_datetime(customer_wrangled['effective_to_date'])
customer_wrangled['month'] = customer_wrangled['effective_to_date'].dt.month
customer_wrangled['week'] = customer_wrangled['effective_to_date'].dt.isocalendar().week
customer_wrangled['day'] = customer_wrangled['effective_to_date'].dt.day
```

```python
for column in customer_wrangled.columns:
    if customer_wrangled[column].dtype == 'object':
        customer_wrangled[column] = pd.factorize(customer_wrangled[column])[0]
```
